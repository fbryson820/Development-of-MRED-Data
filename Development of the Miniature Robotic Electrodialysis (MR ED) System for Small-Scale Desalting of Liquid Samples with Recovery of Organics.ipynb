## Test Data Visualization <a class="anchor" id="Code"></a>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extractData(FILENAME):
    FILENAME = "Data/" + FILENAME
    x_array = []
    y_array = []
    z_array = []
    with open(FILENAME, "r") as filestream:
        for line in filestream:
            currentline = line.split(",")
            x_array.append(float(currentline[0]))
            y_array.append(float(currentline[1]))
            try:
                z_array.append(float(currentline[2]))
            except:
                pass
    return x_array, y_array, z_array

def plotConductivityData(FILENAME):
    t, c, T = extractData(FILENAME)
    plt.plot(t,c,'k.')
    plt.xlabel("Time [s]")
    plt.ylabel("Conductivity [uS/cm]")

def plotPowerData(FILENAME):
    t, c, v = extractData(FILENAME)
    if len(v)==0:
        plt.plot(t,c,'k.')
        plt.xlabel("Time [s]")
        plt.ylabel("Current [mA]")
    else:
        plt.figure()
        plt.subplot(2,1,1)
        plt.plot(t,c,'k.')
        plt.xlabel("Time [s]")
        plt.ylabel("Current [mA]")

        plt.subplot(2,1,2)
        plt.plot(t,v,'k.')
        plt.xlabel("Time [s]")
        plt.ylabel("Voltage [V]")
```
